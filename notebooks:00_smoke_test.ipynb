```python
import torch

print("torch:", torch.__version__)
mps_available = torch.backends.mps.is_available()
mps_built = torch.backends.mps.is_built()
print("MPS built:", mps_built)
print("MPS available:", mps_available)

device = "mps" if mps_available else "cpu"
print("Using device:", device)

x = torch.randn(1024, 1024, device=device)
y = x @ x.T
print("OK, y mean:", y.mean().item())
```

```python
import cv2
import numpy as np
import mediapipe as mp

print("cv2:", cv2.__version__)
print("mediapipe:", mp.__version__)

# создаём тестовую картинку (просто чёрный кадр)
bgr = np.zeros((480, 640, 3), dtype=np.uint8)

# MediaPipe ожидает RGB
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

hands = mp.solutions.hands.Hands(
    static_image_mode=True,
    max_num_hands=2,
    min_detection_confidence=0.5
)

result = hands.process(rgb)
hands.close()

print("Hands landmarks:", result.multi_hand_landmarks)
print("OK: MediaPipe ran successfully")
```
